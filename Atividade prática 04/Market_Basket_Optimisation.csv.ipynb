```python
import pandas as pd 
import numpy as np 
from apyori import apriori 

dataset = pd.read_csv("Market_Basket_Optimisation.csv",header=None)
```

```python
dataset
```

```python
dataset.shape
```

```python
records = []

for i in range(0,7501):
    records.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
#Building the apriori model
rules = apriori(records,min_support=0.05,min_confidence=0.20,min_lift=1.1,min_length=2)
rules_result = list(rules)
```

```python
#Getting the number of rules
print(len(rules_result))
```

```python
#Showing association rules
for i in range(0,len(rules_result)):
    print(rules_result[i],'\n')
```
